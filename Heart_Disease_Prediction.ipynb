```python
# Import Essential Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,r2_score
```

```python
#  Load the Dataset
df=pd.read_csv("heartdataset.csv")
df.head()
```

```python
# Checking Missing Values
df.isnull().sum()
```

```python
# Shape of the Dataset
df.shape
```

```python
#  Statistical summary of the Dataset
df.describe()
```

```python
#  Separating the Dataset into Dependent and Independent Variable
X=df.drop(["target","sex","fasting blood sugar","resting ecg"],axis=1) #Dependent Variables
y=df["target"] # Independent Variables
```

```python
X.head(2)
```

```python
y
```

```python
#  Splitting the dataset into trian_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#  Applying RandomForestClassifier Model
rfc=RandomForestClassifier(n_estimators=100,random_state=0)
rfc.fit(X_train,y_train)
```

```python
# Accuracy on trianing  data
X_train_pred=rfc.predict(X_train)
training_data_accuracy=accuracy_score(X_train_pred,y_train)
training_data_accuracy
```

```python
#  Accuracy on test data
X_test_pred=rfc.predict(X_test)
test_data_accuracy=accuracy_score(X_test_pred,y_test)
test_data_accuracy
```

```python
#  Testing the prediction of the model
input_data=(49,3,160,180,156,0,1.2,0)
input_data_as_numpy_array=np.asarray(input_data).reshape(1,-1)
prediction=rfc.predict(input_data_as_numpy_array)
print(prediction)
if (prediction[0]==0):
    print("The Person is free from Heart Disease")
else:
    print("The Person is suffering from Heart Disease")
```
